```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
file_path = 'data2.csv'  # Adjust path if necessary
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataframe
print("First few rows of the dataset:")
print(df.head())
```

```python

# Display summary statistics
print("\nSummary statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Plotting pairwise relationships to understand correlations
plt.figure(figsize=(10, 8))
sns.pairplot(df, diag_kind='kde')
plt.suptitle("Pairwise Relationships in the Dataset", y=1.02)
plt.show()
```

```python
# Filter out non-numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Display correlation heatmap for numeric features only
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
